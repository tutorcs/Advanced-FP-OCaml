## 1. Basics: what is a function?

Let us begin with a couple of examples of functions we know, namely ‚Äúmathematical functions‚Äù, and then generalize the notion of a function.

<span class="example">Example 1</span> You are probably already familiar with $x \mapsto \sin(x)$




This function, as many other functions you know ($x \mapsto -x$, or $x \mapsto \tanh(x)$, or ...), takes a single number as input, and outputs a single number. In this course, we will describe this using the following notation:

```
sin : number -> number
```

<span class="example">Example 2</span> The square root function: $x \mapsto \sqrt{x}$.

<div>

This one only accepts _positive_ inputs, and returns positive outputs, so we can write:
```
sqrt : positive_number -> positive_number
```

<span class="example">Example 3</span> The modulus of a complex number, $z \mapsto | z |$

<div></div>

This function takes a complex number, and outputs a positive real number:

```
modulus: complex_number -> positive_number
```

Now, a complex number can be represented as a pair of real numbers (real and imaginary parts). So we could just as well write this as:
```
modulus: (number * number) -> positive_number
```

<span class="example">Example 4</span> As a final example, consider the function that converts a complex number expressed in Cartesian coordinates, to its corresponding polar representation. This function takes a pair $(x,y)$, and returns a pair $(r, \theta)$: 

 ```
 cartesian_to_polar : (number * number) -> (positive_number * number)
 ```

We could go on and on, giving yet more examples of functions that you know. What do they have in common?

- They have an <u>input domain</u>, i.e. a set of acceptable inputs, or ‚Äúarguments‚Äù
- They have an <u>output domain</u> (sometimes called ‚Äúco-domain‚Äù in mathematics), which is the kind of things they compute and return
- They <u>do NOT modify the state of your world</u>. Evaluating `sin(x)` cannot change anything else in your program. This might sound silly, but this feature called ‚Äúimmutability‚Äù is a hallmark of functional programming and really sets this paradigm apart from so-called ‚Äúimperative‚Äù programming (e.g. C, python, ...).

Note that the transformation from input to output can be arbitrarily complex!  
<div></div>

For example, you could consider the function that takes a picture, and returns the name and age of each person in the picture. This function would have the form of
```
f : image -> list_of (text * integer_number)
```

<div class="emph_big">Key message: functional programming is all about writing functions and composing them into more complex functions.</div>

In this course, we will begin by writing really simple functions, which are in fact readily available as part of the <a href="https://ocaml.org/api/index.html">standard library</a> or other <a href="https://opam.ocaml.org">third-party libraries</a>. We will then progress to writing more sophisticated algorithms. Bear in mind, though, that it is in the <a href="dev.realworldocaml.org">real world</a> that  OCaml shines, as it enables robust and massively scalable software design in areas as diverse as web front-ends (e.g. Facebook messenger) and backends (web servers, databases), systems (scripting, unikernels, etc), <a href="https://ocaml.xyz">scientific computing</a> (e.g. in <a href="https://cbl-cambridge.org/hennequin">my lab</a>), crypto-currencies (e.g. Tezos), finance, formal logic (e.g. theorem proving), meta-programming (e.g. reasoning about programs), and many other areas!

For function composition to be an effective way of expressing computations, we need to be able to specify complex input and output domains, which programmers also call TYPES. Before we are ready to learn about functions in more depth,  we first need to learn more about types.

## 3. Let-bindings: giving names to values.

### 3.1 Top-level bindings

In order to build more interesting programs, we need to be able to give names to values, otherwise we simply cannot re-use them! OCaml does this through so-called <span class="emph">let bindings</span>:

````ocaml
let x = true
````

Notice how OCaml now tells us `val x : [...]` -- this means ‚ÄúI now hold a <u>val</u>ue called `x` in memory‚Äù.

You can override any let binding previously established:

````ocaml
let x = 5
````

The name `x` has now be re-bound to value `5` (of type int) and so now the name `x` no longer denotes the value `true`.

The point of let-bindings is to be able to reuse previously computed values in other expressions. For example, we can re-use the `x` value in this new expression:

````ocaml
let y = 2 * x
````

The `*` operator denotes integer multiplication (more on this later!).
`2 * x` is a simple example of an <i>expression</i> which OCaml can evaluate (here, giving a value of `10`).

## 4. Boolean operators and conditional expressions

Recall: a value of type `bool` can only be `true` or `false`. Much of the internal logic of computer programs relies on the ability to perform operations on values of type `bool` using __not__, __and__, __or__, and the ability to construct <span class="emph">conditional expressions</span>.   

````ocaml
let a = true
let b = not a
````

````ocaml
(* || means "or" *)
let c = if a || b then 5 else 6
````

````ocaml
(* && means "and" *)
let d = if a && b then 3 else 4
````

Notice how `if [expr1] then [expr2] else [expr3]` requires:
- `expr1` to be of type `bool` (i.e. `true` or `false`)
- `expr2` and `expr3` to have the _same_ type. This is because otherwise, OCaml wouldn't know what type to attribute to the result, before actually evaluating it. OCaml is a so-called <span class="emph">statically typed language</span>, meaning that the type of any value must be determined statically at compile time, not dynamically at run time. 

Thus, OCaml will complain with useful error messages if you make those mistakes:

````ocaml
if 5 = 6 then "three" else "something else"
````

````ocaml
if true then true else "false" 
````

Static typing, coupled with type inference and polymorphism (more on this later!) is what gives OCaml programmers a very strong sense of ‚Äúif it compiles, it is likely going to do what I intend‚Äù.

### 5.1 Defining simple functions

Now that we know how to manipulate expressions and make sense of their types, we are ready to learn about <span class="emph">functions</span>.

Now that we have defined the `double` function, we can actually use it!

<span class="exo">Exercise  5.1/A</span>
    Write a function <code>f</code> that takes two arguments, and returns the sum of their squares.

One obvious solution is to use, in fact, a single argument, of the ‚Äúpair‚Äù type:

````ocaml
let f (x, y) = (x * x) + (y * y)
````

Here, `(x,y)` is a value of type `int * int`, and is considered a single argument.

But Ocaml has a better way of dealing with multiple arguments:

````ocaml
let f_better x y = (x * x) + (y * y) (* or let f_better = fun x y -> ... *)
````

This function has type `int -> int -> int`. This is quite different from `(int * int) -> int`! But this enables a super useful feature of most functional programming language: <span class="emph">partial evaluation</span>. Observe the following:

````ocaml
f_better 2 3  (* this should be 2*2 + 3*3 = 4 * 9 = 13 *)
````

````ocaml
let g = f_better 2
````

Here `g` corresponds to a _partial evaluation_ of the function `f_better`: it is still awaiting the second argument! Hence the `int -> int` type.

````ocaml
g 3
````

This behaviour will prove useful in many ways. In particular, partial evaluation facilitates a lot of useful optimizations. Imagine that we want to compute `(x * x) + (y * y)` for a single value `x=2` and a million different values `y = 1, 2, 3, ...., 1_000_000`. How many multiplications would you perform if you were to use `f`? The answer is two million.

Consider this slight tweak to `f_better` instead:

````ocaml
let f_better x =
  let x_squared = x * x in
  fun y -> x_squared + (y * y)
````

````ocaml
let g = f_better 3
````

This function computes the same as before, but the partial evaluation `let g = f_better 2` allows us to compute `x_squared = x * x` once and for all. By applying `g` to the million different values of `y`, we are now performing only 1_000_001 multiplications, instead of 2_000_000. Thus, we will be done roughly twice as fast.

Note that such an optimization would be more difficult if we had used `let f (x, y) = [...]`, which cannot be partially evaluted. Here, OCaml has an edge over many other programming languages.

<div class="emph_big">
<u>Pro Tip</u> ‚îÄ
     When writing functions of multiple arguments, think carefully about potential partial evaluation use cases, and ‚Äúpull out‚Äù whatever computation does not depend on later arguments (above, we ‚Äúpulled out‚Äù the computation of <code>x * x</code>).
</div>

### 5.2 Pure vs. impure functions

So far, we have been learning about so-called <span class="emph">pure functions</span>, which take their inputs, compute their outputs, and do not modify anything else about your program whatsoever. How could they? We haven't even introduced the notion of a ‚Äúvariable‚Äù that could be modified. All the values we have manipulated so far are called <span class="emph">immutable</span> values, i.e. you can give them a name (and rename them at the top level) but they cannot be modified internally. For example:

````ocaml
let a = 1
let add_one x = x + a
let a = 3
let result = add_one 5
````

The function `add_one` uses the let binding `let a = 1` and forms a ‚Äúclosure‚Äù in which `a` is replaced by `1`. Re-using the name `a` to denote `3` does not modify the value of `a` in that closure. Hence, the result is 6.

OCaml is a hybrid language which does allow you to switch to a more imperative way of programming when convenient, and so in principle you are allowed to use <span class="emph">mutable data</span> types such as a <span class="emph">reference</span> (which C programmers would call a ‚Äúpointer‚Äù):

````ocaml
let r = ref 0
````

````ocaml
!r
````

(don't worry too much about the `{...}` notation, we'll learn about this later). Here, `r` is an `int ref`, which you should understand as a box that contains a `int` value, which can be replaced by another int value. This can be done via the `:=` operator:

````ocaml
r := 100
````

To read what's ‚Äúin the box‚Äù, use the `!` notation:

````ocaml
!r
````

<div class="emph_big">Although OCaml does let you use mutable data, for pedagogical reasons we will KEEP AWAY from this (almost) entirely in this course! At times, you will find that some programs are perhaps more easily expressed using imperative programming, so bear in mind that OCaml allows you to switch if you ever need to in the future.</div> 

Some OCaml functions have `unit` return type. For example, the `init` function in the `Random` module of the standard library has this type:

````ocaml
Random.init
````

In other words, this is a function that takes an int, and returns nothing ü§î. What use could one have for such a function?

Such a function is likely ‚Äúimpure‚Äù, i.e. it is a function that modifies the state of the world. For example:

````ocaml
let r = ref "Hello"
let f () = r := "Hi"
````

````ocaml
!r
````

````ocaml
f ()
````

Here, `f ()` returns nothing (unit), but modifies the state of the world, by modifying what's in the `r` box...

````ocaml
!r
````

You can imagine how such ‚Äúimpure‚Äù functions can create a big mess in your code: if a function is allowed to change other parts of your programme, you need to be very careful about keeping track of these modifications, or else things can go very wrong. It's even worse if you are using third-party library functions that you did not even write yourself, so you are not even sure what they might modify. And it's getting _really bad_ when  using mutable data in the context of a multi-threaded parallel programme, when different processes might make these modifications asynchronously in an order that you can't easily control.   
<br>

<div class="emph_big">Pure functional programming frees you from having to think of modifications made to values you hold in memory. It eliminates a very large number of potential bugs.</div>

### 5.3 Eager evaluation 

In order to reason about the execution of your program, it's important to know _the order in which things get evaluated_. OCaml has a very simple model, based on:
- <span class="emph">eager evaluation</span> ‚îÄ expressions get evaluated directly, in the order in which they are written
- <span class="emph">arguments passed by value</span> ‚îÄ arguments passed to functions are evaluated _before_ the function gets called. 

````ocaml
let x =
  let _m1 = Fp.Utils.print_msg "First" in
  let _m2 = Fp.Utils.print_msg "Second" in
  5 + 2
````

Here, `_m1` gets evaluated first (to `()`), then `_m2` gets evaluted (to `()` as well), and finally `5+2` (to `int`, which is therefore the type of `x`.

Note: it really looks a bit dumb to write `let _m1 = [...]` when the return type is `unit`, because there is no point giving `()` another name! So when chaining expressions that have `unit` type, OCaml lets you use `;` instead, which greatly simplifies the notation: 

````ocaml
let x =
  Fp.Utils.print_msg "First";
  Fp.Utils.print_msg "Second";
  ();
  5 + 2
````

An expression of the form `x1; x2; x3; ...; xN` evalutes to the last element of the chain (here, `xN`). OCaml will rightly warn you if any of the previous expressions (`x1`, `x2`, ..., or `x(N-1)` does not have type `unit`. Try to evaluate this: `5; false; ()`. 

Now, look at this piece of code:

````ocaml
let square x =
  Fp.Utils.print_msg "squaring";
  x * x

let result =
  square
    (Fp.Utils.print_msg "argument";
     5)
````

````ocaml
let rec apply_n_times n f x = if n = 0 then x else apply_n_times (n - 1) f (f x)
````
